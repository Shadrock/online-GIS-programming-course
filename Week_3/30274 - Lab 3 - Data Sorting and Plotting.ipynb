```python
#Importing Pandas library
import pandas as pd
```

```python
#reading data using pandas
data = pd.read_csv("us_counties_covid19_daily.csv")
```

```python
data.head()
```

```python
#groupby number of cases by state to check the total cases by state
grouped_single = data.groupby('state').agg({'cases': ['sum']})
```

```python
#printing groupby data
print(grouped_single)
```

```python
#renaming column name of number of cases
grouped_single.columns= ['Total_Cases']
```

```python
#indexing the rows
grouped_single = grouped_single.reset_index()
```

```python
#printing with the new column name
print(grouped_single)
```

```python
#sorting the data from descending to ascending
grouped_single.sort_values(by=['Total_Cases'], ascending=False)
```

```python
#filtering the original data by state
filter_state = data[(data.state == "New York")]
```

```python
#printing the filered data
print(filter_state)
```

```python
#exporting the filtered data in csv format
filter_state.to_csv("concat_state.csv")
```
